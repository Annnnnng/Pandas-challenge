```python
import pandas as pd
```

```python
# Read the csv file and display data
df = pd.read_csv("Resources/purchase_data.csv")
df
```

```python
# 1. Payler Count
```

```python
#Calculate total number of players
total_players = len(df["SN"].value_counts())
total_players
```

```python
#Display total number of players
df["Total Players"] = total_players
df.loc[[0], ["Total Players"]]
```

```python
# 2.Purchasing Analysis (Total)
```

```python
#Calculate number of unique items
num_of_u_items = len(df["Item Name"].value_counts())
#Display number of unique items
df["Number of Unique Items"] = num_of_u_items
df.loc[[0], ["Number of Unique Items"]]
```

```python
#Calculate average price
ave_price = df["Price"].sum()/df["Price"].count()
# Display average price
df["Average Price"] = ave_price
df["Average Price"] = df["Average Price"].map("${:.2f}".format)
df.loc[[0], ["Average Price"]]
```

```python
#Calculate number of purchases
num_of_purchase = df["Purchase ID"].count()
#Display number of purchases
df["Number of Purchases"] = num_of_purchase
df.loc[[0], ["Number of Purchases"]]
```

```python
#Calculate total revenue
total_rev = df["Price"].sum()
#Display total revenue
df["Total Revenue"] = total_rev
df["Total Revenue"] = df["Total Revenue"].map("${:.2f}".format)
df.loc[[0], ["Total Revenue"]]
```

```python
#Finale result. Display Purchasing Analysis (Total) by combining the columns together
df[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
Purchasing_Analysis = df.loc[[0], ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
Purchasing_Analysis
```

```python
# 3.Gender Demographics
```

```python
#Remove duplicate players with same "SN"
grouped_df = df.drop_duplicates(subset = ["SN"])
#Display grouped data
grouped_df.head()
```

```python
#Calculate the count of each gender
gender_count = grouped_df["Gender"].value_counts()
gender_count
```

```python
#Calculate the percentage of players
percent_of_players = (gender_count / total_players).map("{:,.2%}".format)
percent_of_players
```

```python
# Final result. Display the Gender Demographics table
gender_summary_df = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": percent_of_players})
gender_summary_df
```

```python
# 4. Purchasing Analysis(Gender)
```

```python
#Group data by gender
gender_df = df.groupby(["Gender"])
gender_df
gender_df.count()
```

```python
# Calculate total purchase value
summary_df = pd.DataFrame(gender_df["Price"].sum())
summary_df
```

```python
#Calculate purchase count
purchase_count = gender_df["Purchase ID"].count()
#Display purchase count Column
summary_df["Purchase Count"] = purchase_count
summary_df[["Purchase Count"]]
```

```python
#Calulate average purchase price
ave_pur_price = summary_df["Price"]/summary_df["Purchase Count"]
ave_pur_price
```

```python
#Use map to format average purchase column
summary_df["Average Purchase Price"] = ave_pur_price
summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:.2f}".format)
summary_df[["Average Purchase Price"]]
```

```python
#Calculate total purchase value
pur_value = gender_df["Price"].sum()
summary_df["Total Purchase Value"] = pur_value
#Use map to format total purchase value column
summary_df["Total Purchase Value"] = summary_df["Total Purchase Value"].map("${:.2f}".format)
summary_df[["Total Purchase Value"]]
```

```python
#Calculate average total purchase per person
ave_purchase = (pur_value / gender_count).map("${:.2f}".format)
ave_purchase
```

```python
# Print avg total purchase per person column
summary_df["Avg Total Purchase per Person"] = ave_purchase
summary_df[["Avg Total Purchase per Person"]]
```

```python
#Final result. Display the purchasing analysis by gender by putting all the columns together
summary_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
```

```python
# 5. Age Demographics
```

```python
# Remove duplicate SN
age_df = df.drop_duplicates(subset = ["SN"])
age_df
```

```python
#Create bins
bins = [0, 9.9, 14, 19, 24, 29, 34, 39, 100]
#Create group names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Group players by age
age_df["Age_Ranges"] = pd.cut(age_df["Age"], bins, labels = group_names, include_lowest = True)
age_df["Age_Ranges"].value_counts()
```

```python
#Calculate number of players fall into each age range
total_count = age_df["Age_Ranges"].value_counts()
total_count
```

```python
#Calculate percentage of players by age ranges
percentage_of_players = (total_count / total_players).map("{:,.2%}".format)
percentage_of_players
```

```python
#Final Result. Display the Age Demographics
age_summary_df = pd.DataFrame({"Total Count": total_count, "Percentage of Players": percentage_of_players})
age_summary_df.sort_index()
```

```python
# 7.Top Spenders
```

```python
# Group data by SN
spender_df = df.groupby(["SN"])
spender_df.count()
```

```python
# Calculate purchase count per SN
purchase_count = df["SN"].value_counts()
purchase_count
```

```python
# Calculate total purchase value per SN
total_value = spender_df["Price"].sum()
total_value
```

```python
# Calculate average purchase value per SN
ave_purchase = (total_value / purchase_count).map("${:.2f}".format)
ave_purchase
```

```python
# Print purchase count, average purchase price and total purchase value by each SN
purchase_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": ave_purchase, "Total Purchase Value":
                              total_value})
purchase_df.head()
```

```python
# Final Result. Dislay Top Spenders table bt sorting total purchase value 
top_spender_df = purchase_df.sort_values("Total Purchase Value", ascending=False)
top_spender_df.head().style.format({"Total Purchase Value":"${:.2f}"}) 
```
